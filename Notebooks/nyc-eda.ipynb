```python
# Built-in module:
from os import getcwd, listdir, path
import sqlite3
from time import perf_counter
from copy import deepcopy

# Data Manipulation modules:
import pandas as pd
import numpy as np


# Visualization
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.style import use

# use('ggplot')
```

## Setting paths:

```python
NB_PATH = getcwd()
BASE = path.abspath(path.join(NB_PATH, '..'))
DATA = path.join(BASE, 'Data')
NYC_FP = path.join(DATA, 'NYC-TAXI-TRIP-CLEANED.parquet')
DF = pd.read_parquet(NYC_FP)
```

```python
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=9)
plt.rc('ytick', labelsize=9)

DF.select_dtypes(include=np.number).hist(bins=20, figsize=(16,15), density=True)
plt.show()
```

```python

q99 = DF.trip_duration.loc[(DF.trip_duration >= 0)].quantile(q=.999)
print(q99)
```

```python
DF = deepcopy(
    (
        DF
        .loc[(DF.trip_duration >= 0) & (DF.trip_duration <= 140) & (DF.trip_distance >= 0) & (DF.trip_distance <= 140)]        
    )
)
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=9)
plt.rc('ytick', labelsize=9)

DF.select_dtypes(include=np.number).hist(bins=20, figsize=(16,15), density=True)
plt.show()
```
